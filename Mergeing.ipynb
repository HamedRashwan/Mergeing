```python
customers_orders_df=df[['customer_id','order_id']].drop_duplicates()
df_customer_ord=customers_orders_df.groupby('customer_id').order_id.count()\
.reset_index()
df_customers_cntorders=df_customer_ord.sort_values('order_id',ascending=False)\
.reset_index().drop('index',axis=1)
df_customers_cntorders.columns=['customer_id','count_orders']
df_customers_cntorders
```

```python
def counts_by(col1,col2,v=5):
    
    '''
    col1: idx
    col2: categorical feature
    
    '''
    return   list(df[[col1,col2]].drop_duplicates()[col2].value_counts().head(v).index)
```

```python
d=counts_by('customer_id','city',4)
d
```
